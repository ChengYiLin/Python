## <font color="blue">Read Image with python </font>

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# use PIL(Python Image Library) to processing picture

from PIL import Image
```

### open Image

```python
# read the Image file

pic = Image.open("img/dog.jpg")
pic
```

```python
type(pic)
```

### transform picture to array

```python
pic_arr = np.asarray(pic)
type(pic_arr)
```

```python
# 

pic_arr.shape
```

```python
# show the picture from array_data

plt.imshow(pic_arr)
```

```python
# Let's copy the array for us to use

pic_red = np.copy(pic_arr)
plt.imshow(pic_red)
```

```python
pic_red.shape
```

```python
#  0,1,2  R, G, B 

plt.imshow(pic_red[:,:,0])
```
