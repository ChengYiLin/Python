## <font color="blue">Read Image with python </font>

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# use PIL(Python Image Library) to processing picture

from PIL import Image
```

### open Image

```python
# read the Image file

pic = Image.open("img/dog.jpg")
pic
```

```python
type(pic)
```

### transform picture to array

```python
pic_arr = np.asarray(pic)
type(pic_arr)
```

```python
# 查看圖片的大小

pic_arr.shape
```

```python
# show the picture from array_data

plt.imshow(pic_arr)
```

```python
# Let's copy the array for us to use

pic_red = np.copy(pic_arr)
plt.imshow(pic_red)
```

```python
pic_red.shape
```

```python
# 第三維度 0,1,2 分別是 R, G, B 

plt.imshow(pic_red[:,:,0])
```

你看這圖片的顏色有點奇怪，因為這是 matplotlib 預設所使用的顏色，越黃的代表顏色的強度越強，越藍則反之。

像是我們的這張是 R 的部分，越黃代表紅色程度越強，反之亦然。

但如果我們還是不習慣，看黑白色的圖還是比較爽，那就多加一個參數即可

```python
# 紅色階層

plt.imshow(pic_red[:,:,0],cmap='gray')
```

我們來看看其他顏色是不是會和這一張圖變化

```python
# 綠色階層

plt.imshow(pic_red[:,:,1],cmap='gray')
```

```python
# 藍色階層

plt.imshow(pic_red[:,:,2],cmap='gray')
```

但我就是想看全紅的照片啊~ 那要怎麼做呢?

```python
# 我們把 G 給取消掉

pic_red[:,:,1] = 0
```

```python
plt.imshow(pic_red)
```

```python
# 再把 B 給取消掉

pic_red[:,:,2] = 0
```

```python
plt.imshow(pic_red)
```
